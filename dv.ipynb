```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('adidas_usa.csv')
df = data.head(50)
df
```

```python
colors_column = data['color']
color_counts = colors_column.value_counts()
plt.figure(figsize=(10, 6))
color_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Shoes by Color')
plt.xlabel('Color')
plt.ylabel('Number of Shoes')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.bar(df.color,df.selling_price)
```

```python
category_column = data['category']
cate_counts = category_column.value_counts()
plt.figure(figsize=(10, 6))
cate_counts.plot(kind='bar', color='green')
plt.title('Number of Products by Categories')
plt.xlabel('Categories')
plt.ylabel('Number of Shoes')
plt.xticks(rotation=45)
plt.show()
```

```python
category=data['category'].value_counts()
plt.figure(figsize=(3,3))
plt.pie(category,labels=category.index,autopct='%1.1f%%')
plt.title("Category of Items")
plt.axis("equal")
plt.show()
```
